```python
import glob, os, tensorflow, keras_video.utils
import numpy as np
from keras_video import VideoFrameGenerator
from tensorflow import keras

model = keras.models.load_model('/content/drive/MyDrive/tirocinioWorkingDirectory/datasets/grid/weights/2810ConvLSTM2D/weights.38.hdf5')

# some global params
SIZE = (80, 40)
CHANNELS = 1
NBFRAME = 3
BS =  1

#unzip full dataset from gdrive to obtain all classes
!rm -rf "/content/sample_data"
!rm -rf "/content/myFinalDataset"
#!rm -rd "/content/myDataset"
!unzip -qq "/content/drive/MyDrive/tirocinioWorkingDirectory/datasets/grid/finalDataset/myFinalDatasetCroppedNewEdition.zip" -d "/content/fullDataset"
!mv "/content/fullDataset/content/myDataset/" "/content/myFullDataset"
!rm -rf "/content/fullDataset"

#unzip test dataset
!!unzip -qq "/content/drive/MyDrive/tirocinioWorkingDirectory/datasets/grid/finalDataset/myTestDataset.zip" -d "/content/testDataset"
!mv "/content/testDataset/myFinalDataset/" "/content/myTestDataset"
!rm -rf "/content/testDataset"

#class names
fullDatasetPath = os.path.join('/content', 'myFullDataset')
allClasses = [i.split(os.path.sep)[3] for i in glob.glob(os.path.join(fullDatasetPath, '*'))]
allClasses.sort()
print(allClasses)

testDatasetPath = os.path.join('/content', 'myTestDataset')
testClasses = [i.split(os.path.sep)[3] for i in glob.glob(os.path.join(testDatasetPath, '*'))]
testClasses.sort()
print(testClasses)
```

```python
tot = 0
for aClass in testClasses:
  print("Class: ", aClass)
  path = os.path.join('/content', 'myTestDataset', aClass)
  # Create video frame generator
  test = VideoFrameGenerator(
  classes=[aClass], 
  glob_pattern=os.path.join('/content', 'myTestDataset', '{classname}', '*'),
  nb_frames=NBFRAME,
  target_shape=SIZE,
  nb_channel=CHANNELS,
  use_frame_cache=False,
  batch_size=1
  )
  #predict
  predictions = model.predict(x=test)
  #view predictions
  #predictedClasses = np.argmax(predictions, axis = 1)
  #predictedWord = [allClasses[i] for i in predictedClasses]
  n = 5
  for j in range(len(predictions)):
    indices = (-predictions[j]).argsort()[:n]
    predictedWords = [allClasses[i] for i in indices]
    print("Predicted: ", predictedWords)
    if aClass in predictedWords:
      tot = tot + 1
  print("\n")

print("Results on top-", n, ":", sep="")
print("Predicted: ", tot, "on 20")
print("Accuracy percentage: ", tot*100/20, "%", sep="")
```
